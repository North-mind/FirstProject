```python
import pickle
from sklearn.model_selection import train_test_split
```

```python
with open('../newsDataPP.pickle', 'rb') as pickleInput:
    articlesData = pickle.load(pickleInput)

articlesData.head(2)
```

```python
articlesData['gensimCleaner'] = articlesData['gensimCleaner'].apply(lambda content: " ".join([word for word in content])) 
articlesData['disciplineName'].value_counts()
```

```python
articlesData['disciplineName'].value_counts()
articlesData = articlesData[~articlesData['disciplineName'].isin(['Snooker', 'NFL'])]
```

```python

X = articlesData.gensimCleaner
y = articlesData.disciplineName

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 30)
```

```python
disciplines = list(y_test.unique())
len(disciplines)
```

```python
#Naive Bayes Classifier for Multinomial Models

from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import classification_report

nb = Pipeline([('vect', CountVectorizer()),
              ('tfidf', TfidfTransformer()),
              ('clf', MultinomialNB()),
              ])

%time nb.fit(X_train, y_train)
y_pred = nb.predict(X_test)

print('accuracy {}'.format(accuracy_score(y_test, y_pred)))
print('Report:\n', classification_report(y_test, y_pred, target_names = disciplines))
#print(confusion_matrix(y_test, y_pred))
```

```python
list(y_pred)
```

```python
# Linear Support Vector Machine
from sklearn.linear_model import SGDClassifier

nb = Pipeline([('vect', CountVectorizer()),
              ('tfidf', TfidfTransformer()),
              ('clf', SGDClassifier())
              ])
%time nb.fit(X_train, y_train)
y_pred = nb.predict(X_test)

print('accuracy {}'.format(accuracy_score(y_test, y_pred)))
print('Report:\n', classification_report(y_test, y_pred, target_names = disciplines))
#print(confusion_matrix(y_test, y_pred))
```
